```python
import os

demo_file = '/content/drive/MyDrive/SadTalker_Patched_2026/src/gradio_demo.py'

if os.path.exists(demo_file):
    with open(demo_file, 'r') as f:
        content = f.read()

    # We create a specific 'display' version in a different folder
    # so Gradio doesn't try to overwrite the source file in 'results'
    old_logic = "return final_output"
    new_logic = """
        web_display_path = os.path.join('/content', 'gradio_display_output.mp4')
        shutil.copy(final_output, web_display_path)
        return web_display_path
    """

    if old_logic in content:
        content = content.replace(old_logic, new_logic)
        with open(demo_file, 'w') as f:
            f.write(content)
        print("✅ Gradio Display Redirected. The UI should now stay stable.")
    else:
        print("⚠️ Patch target not found. Check if logic was already updated.")
```

```python
import os
import shutil

demo_file = '/content/drive/MyDrive/SadTalker_Patched_2026/src/gradio_demo.py'

def apply_final_patch():
    if not os.path.exists(demo_file):
        print("❌ File not found. Check Drive mount.")
        return

    with open(demo_file, 'r') as f:
        content = f.read()

    # The most reliable way to find the return line
    # We look for the shutil.copy line we added earlier and update it
    if "return final_output" in content:
        new_logic = """        web_display_path = os.path.join('/content', 'gradio_display_output.mp4')
        if os.path.exists(final_output):
            import shutil
            shutil.copy(final_output, web_display_path)
        return web_display_path"""

        updated_content = content.replace("        return final_output", new_logic)

        with open(demo_file, 'w') as f:
            f.write(updated_content)
        print("✅ SUCCESS: Gradio UI display redirected to /content.")
    elif "web_display_path" in content:
        print("ℹ️ File already patched.")
    else:
        print("⚠️ Could not find the return line. Manual check required.")

apply_final_patch()
```

```python
import os

app_file = '/content/drive/MyDrive/SadTalker_Patched_2026/app_sadtalker.py'

if os.path.exists(app_file):
    with open(app_file, 'r') as f:
        content = f.read()

    # We add allowed_paths=["/content"] to the launch command
    if '.launch(' in content and 'allowed_paths' not in content:
        content = content.replace('.launch(', '.launch(allowed_paths=["/content"], ')
        with open(app_file, 'w') as f:
            f.write(content)
        print("✅ Gradio Launch Security updated: /content is now allowed.")
    else:
        print("ℹ️ Launch command already updated or not found.")
```
